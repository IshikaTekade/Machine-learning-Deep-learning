# Linear Regression Model
## function(f) where f(x)=wx+b

### x= training example features
### y= training example targets
### x^i,y^i =i^th training example
### m=number of training examples


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#x train will be the input features which is size(1000 sq.ft)
# y train will be the actual or target variable (1000 dollars)
x_train=np.array([1.0,2.0])
y_train=np.array([300.0,500.0])
x_train,y_train
```

```python
x_train.shape
```

```python
m=x_train.shape[0]
m
```

```python
#x_i y_i
i=0
x_i=x_train[0]
y_i=y_train[0]

print(f"(x^{i}),(y^{i})=({x_i},{y_i})")
```

```python
#plotting the data
plt.scatter(x_train,y_train,marker="x", c="r")
plt.title("HOUSING PRICES")
plt.xlabel("Size(1000 sq.ft)")
plt.ylabel("Price ($1000)")
plt.show()

```

```python
#Model Function
#f(x(i)=wx(i)+b)
w=200
b=100
x_train[0]

#So basically
#x[0]; f(x)=w*x[0]+b and so on for x[1]..
#we will therefore use a for loop
```

```python
def compute_model_output(x,w,b):
    m=x.shape[0]
    f_x=np.zeros(m) #creates a 1d array of m(2) entries
    for i in range(m):
        f_x[i]=w*x[i]+b
    return f_x
```

```python
tmp_f_wb=compute_model_output(x_train,w,b)
tmp_f_wb
```

```python
#plot our model prediction
plt.plot(x_train,tmp_f_wb,c="b",label="Our Prediction")
plt.scatter(x_train,y_train,marker="x",c="r",label="Actual Values")
plt.title("Housing Prices")
plt.ylabel("Price ")
plt.xlabel("Size")
plt.show()
```

## Example 2 (Predicting car price with their age)


```python
#Dataset x1_train has the age of the car(years)
# y1_train has the price of the car(in 1000's)

x1_train=np.array([2,3,5])
y1_train=np.array([40,35,28])
x1_train,y1_train
```

```python
w=-7.861
b=60.5957
def compute_moutput(x,w,b):
    m=x.shape[0]
    f_x1=np.zeros(m)
    for i in range(m):
        f_x1[i]=w*x[i]+b
    return f_x1
```
